```python
import sqlite3

# Connect to an in-memory database (for demonstration)
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create a sample table
cursor.execute('''
CREATE TABLE products (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    price REAL,
    category TEXT
)
''')
conn.commit()

# Insert some data
cursor.execute("INSERT INTO products (name, price, category) VALUES ('Laptop', 1200, 'Electronics')")
cursor.execute("INSERT INTO products (name, price, category) VALUES ('Keyboard', 75, 'Electronics')")
cursor.execute("INSERT INTO products (name, price, category) VALUES ('Mouse', 25, 'Electronics')")
cursor.execute("INSERT INTO products (name, price, category) VALUES ('Book', 20, 'Books')")
cursor.execute("INSERT INTO products (name, price, category) VALUES ('Notebook', 5, 'Stationery')")
conn.commit()
```

```python
cursor.execute("SELECT * FROM products")
rows = cursor.fetchall()
for row in rows:
    print(row)
```
