```python
fruits = ["apple", "banana", "cherry"]

# For loop
for fruit in fruits:
  print(fruit)

# Accessing elements
print(fruits[0]) #prints apple
print(fruits[-1]) #prints cherry

# List operations
fruits.append("orange")
print(fruits)

fruits.insert(1, "grape") #insert grape at index 1
print(fruits)

fruits.remove("banana")
print(fruits)

print(len(fruits)) #length of list
```
