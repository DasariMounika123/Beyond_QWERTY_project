```python

import pandas as pd

# Create sample data
data = {"Name": ["Alice", "Bob", "Charlie"],
        "Age": [25, 30, 35],
        "Salary": [50000, 60000, 70000]}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("sample_data.csv", index=False)

# Read the CSV file
df_read = pd.read_csv("sample_data.csv")

# Display data
print(df_read)

# Basic statistics
print("\nSummary statistics:")
print(df_read.describe())
```
