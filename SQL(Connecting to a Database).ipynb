```python
import sqlite3

# Connect to an in-memory database (for demonstration)
# If conn was previously defined, close the connection before re-connecting
try:
    conn.close()
except NameError:
    pass # If conn was not defined, we don't need to close it.

conn = sqlite3.connect(':memory:')  # Or connect to a file: sqlite3.connect('my_database.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE employees (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    department TEXT
)
''')
conn.commit()
```
